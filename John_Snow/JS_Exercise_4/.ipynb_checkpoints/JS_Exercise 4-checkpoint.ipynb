# Exercise 5: John Snow's Grand Experiment

#### Datasets: `The Grand Experiment`

Being a doctor, John Snow had seen many people suffer and die from cholera. It primarily affected their gastrointestinal system (vomiting and diarrhea), leading to extreme dehydration and eventually death for many. Therefore, he was convinced that it had something to do with what people were eating or drinking. He focused on the water supplies of the people of South London. However, it would be hard for him to convince people to change their beliefs, but he knew he could do it with the right data.

There were 11 districts in South London. In order to conduct a more detailed investigation, he split these districts into 31 "sub-districts"

The South districts of London were supplied mainly by two water companies: Lambeth and Southwark & Vauxhall. Both companies drew their water from the Thames.



John Snow was able to collect detailed data about the sub-districts and their water supplies. Let's take a look at some of this data.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
Grand_Exp = pd.read_csv("Grand Experiment Data.csv")
Grand_Exp
```

```python
suppliers = Grand_Exp['supplier'].unique()
print(f"List of all suppliers: {suppliers}")
```

John Snow chose to examine the two major outbreaks see if there were differences in how each affected the south districts.

The code below will allow you to calculate the death rate for certain water companies for each of the two outbreaks (1849 and 1854). The way the code is written, it will show you the death rate for people that got their water from Southwark & Vauxhall during the 1849 outbreak. 

**(1) Run the cell, then try changing the variable in the code to get the data for Lambeth.** 

```python
the_mean = Grand_Exp[(Grand_Exp['supplier'] == 'SouthwarkVauxhall')]['rate1849'].mean()
print(f"Mean mortality rate (per 1000): {the_mean}")
```

```python
the_mean = Grand_Exp[(Grand_Exp['supplier'] == 'SouthwarkVauxhall')]['rate1854'].mean()
print(f"Mean mortality rate (per 1000): {the_mean}")
```

**(2) Repeat the process for both water companies for 1854.**

```python
the_mean = Grand_Exp[(Grand_Exp['supplier'] == 'Lambeth')]['rate1849'].mean()
print(f"Mean mortality rate (per 1000): {the_mean}")
```

```python
the_mean = Grand_Exp[(Grand_Exp['supplier'] == 'Lambeth')]['rate1854'].mean()
print(f"Mean mortality rate (per 1000): {the_mean}")
```

Without knowing it, the people of South London had been experimented on between 1849 and 1854. Lambeth had moved its water supply farther upstream on the Thames. Southwark & Vauxhall had not moved their water supply. **Who was the control group and who was the experimental group?**

Notice that a lot of the sub-districts got their water from both companies. John Snow now set his attention on these sub-districts. What advantage did these people have over others in John Snow's Gand Experiment"?

Figuring out who got their water from which company was a big challenge for John Snow because the water pipes in these sub-districts were all mixed up and neighbors could be getting their water from different companies. Often people rented and didn't know who their water company was. 

Fortunately, John Snow was able to figure out a solution by going door to door and testing people's water to determine which water company they had. From this, we have data that describes about what percentage of people in each of these sub-districts got their water from Lambeth (perc_lam) and from Southwark & Vauxhall (perc_sou).

Let's take a look at just these sub-districts:

```python
mixed = Grand_Exp[(Grand_Exp['supplier'] == 'SouthwarkVauxhall_Lambeth')]
mixed
```

**Is there a correlation between water supplier and mortality rate?** 

Let's look at 1849 first. The code below will show you if there is a correlation between getting your water from Lambeth and cholera in 1849. 

```python
x = mixed['perc_lam']
a = mixed['rate1849']


plt.scatter(x, a, color='red')
plt.xlim(right = 1)
plt.ylim(bottom = 0)


m, b = np.polyfit(x, a, 1)
plt.plot(x, m*x + b)

plt.xlabel('Percent Lambeth')
plt.ylabel('Death Rate (per 1000)')
```

**Modify the code above** to see if there is a correlation between getting your water from Southwark & Vauxhall and cholera in 1849.

```python
x = mixed['perc_sou']
a = mixed['rate1849']


plt.scatter(x, a, color='red')
plt.xlim(right = 1)
plt.ylim(bottom = 0)


m, b = np.polyfit(x, a, 1)
plt.plot(x, m*x + b)

plt.xlabel('Percent Southwark')
plt.ylabel('Death Rate (per 1000)')
```

**Now do the same** for 1854 for each water company.

```python
x = mixed['perc_lam']
a = mixed['rate1854']


plt.scatter(x, a, color='red')
plt.xlim(right = 1)
plt.ylim(bottom = 0)


m, b = np.polyfit(x, a, 1)
plt.plot(x, m*x + b)

plt.xlabel('Percent Lambeth')
plt.ylabel('Death Rate (per 1000)')
```

```python
x = mixed['perc_sou']
a = mixed['rate1854']


plt.scatter(x, a, color='red')
plt.xlim(right = 1)
plt.ylim(bottom = 0)


m, b = np.polyfit(x, a, 1)
plt.plot(x, m*x + b)

plt.xlabel('Percent Southwark')
plt.ylabel('Death Rate (per 1000)')
```
