```python
#USE: Calculate Pi using Monte Carlo Method

from ColabTurtle import Turtle as turtle
import random
import math

# initialize turtle graphics
# turtle.setup(600, 600)
# turtle.Screen().setworldcoordinates(-2.0, -2.0, 2.0, 2.0)
t = turtle
t.initializeTurtle(initial_speed = 10, initial_window_size = (300, 300))
t.penup()

num_darts = 1000

# throw `num_darts` darts at random locations within a square
# and check how many land within an inscribed circle
inside = 0
color = 'black'
for i in range(num_darts):
    x = random.randint(0, 300)
    y = random.randint(0, 300)
    if math.sqrt(x ** 2 + y ** 2) <= 300:
        color = 'red'
        inside = inside + 1
    else:
        color = 'grey'
    t.goto(x, y)
    t.color(color)
    t.pendown()
    t.forward(1)
    t.penup()

pi = 4 * inside / num_darts
print(f'pi = {pi}')

#t.done()
```

```python
# Modify: Calculate Pi using Monte Carlo Method Version 2
# Needs some debugging

import random
import numpy

# define arrays
pointx = []
pointy = []

# define the number of points
# more points will be more accurate but slower
total = 40000
halftotal = total/2

# define the points
for x in range(1, total + 1):
    pointx.append(random.randint(1,total))
    pointy.append(random.randint(1,total))

# make the arrays of points mathable
pointy = numpy.array(pointy)
pointx = numpy.array(pointx)

# determine number of points in the circle
incircle = 0
for x in range(1,total):
    if halftotal >= ((halftotal - pointx[x]) ** 2 + (halftotal - pointy[x]) **2) ** 0.5:
        incircle += 1

print(f'total points = {total}')
print(f'in circle = {incircle}')

# determine pi
pi = (4 * incircle) / total
print(f'pi = {pi}')
```
